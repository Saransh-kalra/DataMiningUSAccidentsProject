```python
def categorical_to_numeric(series):
    """
    This takes in a pandas Series and returns and array containing the
    number of occurences of each category
    """
    # Convert the Series into a numpy array
    data_array = np.array(series)
    
    # Exctract the categories
    categories = np.unique(data_array)
    
    # Create an array of zeros to hold the num of occurence
    # for each category
    category_occurences = [0] * len(categories)
    
    for index, category in enumerate(categories):
        for data_point in data_array:
            if data_point == category:
                category_occurences[index] += 1
    
    return list(categories), category_occurences
    
```

```python
def categorical_pie_chart(data, labels, title=None):
    fig1, ax1 = plt.subplots()
    ax1.pie(data, labels=labels, autopct='%1.1f%%')
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    if title is None:
        pass
    else:
        ax1.set(title=title)
    
    plt.show()
```

```python
age_data = dataset_df['age']
categories, category_occurences = categorical_to_numeric(age_data)
categorical_pie_chart(category_occurences, categories, 'Age')
# print(np.array(categories).shape)
# print(np.array(category_occurences).shape)

# plt.hist(np.array(categories), weights=np.array(category_occurences))
```
